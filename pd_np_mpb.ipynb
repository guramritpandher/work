```python
arr = np.array([10, 20, 30, 40])

print(arr + 5)   # [15 25 35 45] (adds 5 to each element)
print(arr * 2)   # [20 40 60 80] (multiplies each element by 2)
print(arr / 10)  # [1. 2. 3. 4.]
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)  # [5 7 9]
print(a * b)  # [4 10 18]
print(a / b)  # [0.25 0.4 0.5]
```

```python
arr = np.array([10, 20, 30, 40, 50])

print(np.mean(arr))  # 30.0 (average)
print(np.median(arr))  # 30.0 (middle value)
print(np.std(arr))  # Standard deviation
print(np.sum(arr))  # 150 (sum of all elements)
print(np.max(arr))  # 50 (max value)
print(np.min(arr))  # 10 (min value)
```

```python
matrix = np.array([[1, 2], [3, 4]])

print(np.linalg.inv(matrix))  # Inverse of a matrix
print(np.dot(matrix, matrix))  # Matrix multiplication
print(np.transpose(matrix))  # Transpose of a matrix
```

# **MatPlotLib**

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])  # X-axis values
y = np.array([10, 20, 25, 30, 50])  # Y-axis values

plt.plot(x, y, marker='o', color='b', label="Sales Trend")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Simple Line Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(50) * 10  # Random X values
y = np.random.rand(50) * 100  # Random Y values

plt.scatter(x, y, color='red', alpha=0.5)
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.title("Random Scatter Plot")
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [10, 30, 20, 40]

plt.bar(categories, values, color=['blue', 'orange', 'green', 'red'])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Simple Bar Chart")
plt.show()
```

```python
data = np.random.randn(1000)  # 1000 random values from normal distribution

plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```
